```python
import pandas as pd
```

```python
df =pd.DataFrame({'col1':[1,2,3,4,5],
              'col2':[45,56,67,34,34] ,
             'col3':"nishit hirpara krish ineuron one".split()
            })
```

```python
df
```

```python
df['col4']=df['col1']**2
```

```python
df
```

```python
df['col5']=df['col3'].str[0]
```

```python
df
```

```python
def test(a):
    return a**2
```

```python
df['col4_4']=df['col1'].apply(test)
```

```python
df
```

```python
def test1(a):
    if a=='n':
        return 'n'
    else:
        return 'z'
        
```

```python
# using external function test1
df['col6']=df['col5'].apply(test1)
```

```python
df
```

```python
# same thing achieve by using lamda function
df['col6_6']=df['col5'].apply(lambda a:'n' if a=='n' else 'z')
```

```python
df
```

```python
#apply len function (inbuilt function)
df['col7']=df['col3'].apply(len)
```

```python
df
```

```python
# apply log of col2 in col8
df['col8']=df['col2'].apply(math.log)
```
